```python
# Label encoding

label = preprocessing.LabelEncoder()
retail_data_encoded = pd.DataFrame()

retail_data_1_col = ['Product Type', 'Item', 'Region Level 1', 'Region Level 2', 'Urban/Rural']
for i in retail_data_1_add0.columns:
    if i in retail_data_1_col:
        retail_data_encoded[i] = label.fit_transform(retail_data_1_add0[i])
    else:
        retail_data_encoded[i] = retail_data_1_add0[i]
```

```python
train_1 = retail_data_encoded[retail_data_encoded['Datetime'] < set_dt]
val_1 = retail_data_encoded[retail_data_encoded['Datetime'] >= set_dt] # 2022
```

```python
# Feature와 Target 분리
target = 'Sales Total'
features = train_1.drop([target,"Datetime"], axis=1).columns

X_train = train_1[features]
y_train = train_1[target]
X_val = val_1[features]
y_val = val_1[target]

print(X_train.shape, y_train.shape)
print(X_val.shape, y_val.shape)
```

```python
# 확인
X_train.head(3)
```

```python
# 파이프라인 생성 및 학습
xgb_basic = make_pipeline(
    SimpleImputer(),
    RobustScaler(),
    XGBRegressor(random_state=1,
                 n_jobs=-1,
                 objective='reg:squarederror',
                 max_depth=5, # 5
                 learning_rate=0.2))

xgb_basic.fit(X_train, y_train);
```

```python
# 예측
y_pred = xgb_basic.predict(X_val)

# prediction값 반올림
y_pred_r =  [round(x) if round(x) >=0 else 0 for x in y_pred]

# 결과 확인

print('mae:  ', mae(y_val, y_pred_r))
print('rmse: ', np.sqrt(mse(y_val, y_pred_r)))

# 수정전
# mae:   0.5681168738391199
# rmse:  1.5965796357571027
```

### hyper parameter

```python
# eval set 생성
pre_processing_part = make_pipeline(
    SimpleImputer(),
    RobustScaler())

X_train_processing = pre_processing_part.fit_transform(X_train)
X_val_processing = pre_processing_part.transform(X_val)

eval_set = [(X_train_processing, y_train),
            (X_val_processing, y_val)]
```

```python
# 모델 생성
xgb_hyper = XGBRegressor(
    random_state=1,
    n_jobs=-1,
    objective='reg:squarederror')

# 사용할 하이퍼 파라미터
dists = {
    'max_depth': [3, 4, 5, 6, 7, 8],
    'min_child_weight': [1, 2, 3, 4],
    'gamma': [0, 0.01, 0.1, 0.2, 0.3],
    'learning_rate': [0.001, 0.01, 0.1, 0.2, 0.3]}

# cv
cv = TimeSeriesSplit(n_splits=5)

# GridSearchCV
reg_xgb_hyper = GridSearchCV(
    xgb_hyper, 
    param_grid=dists,
    cv=cv, 
    scoring='neg_root_mean_squared_error',
    verbose=0,
    n_jobs=-1
)

reg_xgb_hyper.fit(X_train_processing,
                  y_train,
                  eval_set = eval_set,
                  eval_metric='rmse',
                  early_stopping_rounds = 50);
```

```python
# 결과 확인
print('best hyper-parameter: ', reg_xgb_hyper.best_params_)
```

```python
# cv 결과 확인
pd.DataFrame(reg_xgb_hyper.cv_results_).sort_values(by='rank_test_score').head().T
```

```python
# 모델 저장
model_xgb_hyper = reg_xgb_hyper.best_estimator_

# 예측
y_pred = model_xgb_hyper.predict(X_val_processing)

# prediction값 반올림
y_pred_r =  [round(x) if round(x) >=0 else 0 for x in y_pred]

# 결과 정리
print('mae :', round(mae(y_val, y_pred_r), 2))
print('rmse :', round(np.sqrt(mse(y_val, y_pred_r)), 2))
```
